**PARTE 1:PROYECTO API**

Escojo una Api, Open Trivia, que genera preguntas aleatorias en función de varios desplegables, según las opciones que selecciones genera una url aleatoria. Importo las librerías necesarias y llamo a la url con requests.get y  obtengo un fichero jason con 10 preguntas aleatorias sobre temas aleatorios también

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
response = requests.get('https://opentdb.com/api.php?amount=10')
results = response.json()
display(results)
```

El resultado de la petición es un fichero json que contiene un diccionario. Para separar los datos por columnas unicamente utilizo json_normalize y hago undisplay de los datos

```python
data = json_normalize(results, 'results')

display(data)
```

Una vez generadas las preguntas, defino una función que me hace 5 preguntas aleatorias y me ofrece las respuestas posibles.El usuario ha de meter la respuesta correcta y el marcador anota la puntuación

```python
def pregunta_trivial():
    contador=1
    resp_ok=0
    for i in range(5):
        pregunta= data.loc[i]['question']
        print(pregunta)
        respuesta = data.loc[i,:]['incorrect_answers']
        respuesta.append(data.loc[i,:]['correct_answer'])
        print(respuesta)
        usuar_respuesta= input('Respuesta: ')
             
        if usuar_respuesta == data.loc[i,:]['correct_answer']: 
            resp_ok += 1
            contador += 1 
            print('Acertaste, ' "llevas " + str(resp_ok) + " de " +str(contador))
            
        else:
            print(" Sigue intentándolo ... llevas " + str(resp_ok) + " de " + str(contador))                
            contador+=1
        
    print('estadistica:', resp_ok,'/',  contador)
 
pregunta_trivial()
```

Como span-5 es lo que he denominado contenedor y lo que quiero es ir de lo general a lo particular, creo un loop que me de, para cada uno de los diferentes aceites en contenedor:

Decido hacer lo mismo que hice en la tabla inicial de nombre y marca pero añadir una tabla de precios vacía en la que incorporo los precios. Obtengo una lista de precios, pero con la lista no puedo crear un dataframe porque no tiene índice, así que transformo la lista en una serie y  creo un dataframe sólo de precios, Evidentemente el resultado no es el más óptimo pero no consigo separar esos precios de otra manera

```python
container=soup.select('div.span-5 ')
tabla = []
precios=[]


for aceite in container:
    nombre_aceite = aceite.a['title']
    tabla.append(nombre_aceite)
    marca_container=aceite.find_all('div',{'class':'marca'})
    marca=marca_container[0].text.strip()
    tabla.append(marca)
    precio_container= aceite.find_all('div',{'class':'priceContainer'})
    preciob = precio_container[0].text.strip()
    precio = re.findall('\d+', preciob)
    precios.append(precio)

        
print(precios)
```

```python
precios=pd.Series(precios)
tabla_precios=pd.DataFrame(precios,columns=['Precios'])


```

Finalmente hago un merge entre los dos dataframes. Me queda pendiente descifrar cómo separar esos datos

```python
df=pd.merge(tabla_aceite, tabla_precios, left_index=True, right_index=True)
display(df)
```
